```python
def calculate_average_length(dataset):
    total_length = 0
    count = 0

    for example in dataset:
        audio_path = example["audio_path"]
        if isinstance(audio_path, dict):
            audio_path = audio_path.get("path")  # Accessing the path from dictionary

        audio_array = torchaudio.load(audio_path)[0].squeeze()
        audio_length = len(audio_array)



        total_length += audio_length
        count += 1

    average_length = total_length / count

    return average_length
```

```python
def preprocess_function(batch, avg_length):
    input_values_list = []
    labels_list = []

    for example in batch["audio_path"]:
        # Ensure we handle cases where audio_path might be a dictionary
        audio_path = example
        if isinstance(audio_path, dict):
            audio_path = audio_path.get("path")

        # Load and process audio
        speech_array, sampling_rate = torchaudio.load(audio_path)
        resampler = torchaudio.transforms.Resample(sampling_rate, 16000)
        speech_array = resampler(speech_array).squeeze()

        # Convert to tensor if it's a numpy array
        if isinstance(speech_array, np.ndarray):
            speech_array = torch.tensor(speech_array)

        # Truncate if longer than max_length
        if len(speech_array) > avg_length:
            speech_array = speech_array[:avg_length]

        # Pad if shorter
        if len(speech_array) < avg_length:
            padding = avg_length - len(speech_array)
            speech_array = torch.nn.functional.pad(speech_array, (0, padding))

        # Convert audio to input values suitable for the model
        input_values = processor(speech_array.numpy(), sampling_rate=16000).input_values[0]
        input_values_list.append(input_values)

    for text in batch["Bengali_Text"]:
        with processor.as_target_processor():
            labels = processor(text).input_ids
            labels_list.append(labels)

    return {
        "input_values": input_values_list,
        "labels": labels_list,
    }
```

```python
from dataclasses import dataclass
from typing import Any, Dict, List, Union
```

```python
@dataclass
class DataCollatorCTCWithPadding:
    processor: Wav2Vec2Processor
    padding: Union[bool, str] = True

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        # Split inputs and labels
        input_features = [{"input_values": feature["input_values"]} for feature in features]
        label_features = [{"input_ids": feature["labels"]} for feature in features]

        # Pad inputs and labels
        batch = self.processor.pad(
            input_features,
            padding=self.padding,
            return_tensors="pt",
        )

        with self.processor.as_target_processor():
            labels_batch = self.processor.pad(
                label_features,
                padding=self.padding,
                return_tensors="pt",
            )

        # Replace padding with -100 to ignore tokens during loss calculation
        labels_batch["input_ids"][labels_batch["input_ids"] == self.processor.tokenizer.pad_token_id] = -100

        batch["labels"] = labels_batch["input_ids"]

        return batch
```

```python
# Initialize the custom data collator
data_collator = DataCollatorCTCWithPadding(processor=processor)
```

```python
import numpy as np
import evaluate

wer_metric = evaluate.load("wer")

def compute_metrics(pred):
    pred_logits = pred.predictions
    pred_ids = np.argmax(pred_logits, axis=-1)

    pred_str = processor.batch_decode(pred_ids)
    label_str = processor.batch_decode(pred.label_ids, group_tokens=False)

    wer = wer_metric.compute(predictions=pred_str, references=label_str)

    print(f"Predictions: {pred_str[:5]}")  # Print first 5 predictions for debugging
    print(f"References: {label_str[:5]}")  # Print first 5 references for debugging
    print(f"WER: {wer}")

    return {"wer": wer}
```
